
Hipótese:
* Os clientes que mais compram são da região sudeste? 
* Qual o tempo médio de entrega dos pedidos?



> Estados com respectivo número de clientes.

```sql
%sql

SELECT customer_state as Estado, count(*) as Total
FROM olist_dataset.default.clientes_olist
GROUP BY customer_state
ORDER BY total DESC
```


> Tempo médio de entrega dos pedidos.

```sql
%sql

SELECT 
    YEAR(order_purchase_timestamp) AS Ano,
    MONTH(order_purchase_timestamp) AS Mes,
    AVG(
        DATEDIFF(
            order_delivered_customer_date, 
            order_purchase_timestamp
        )
    ) AS AVG_tempo_entrega_dia,
    COUNT(*) AS Total_Pedidos
FROM olist_dataset.default.pedidos_olist
WHERE 
    order_status = 'delivered'
    AND order_delivered_customer_date IS NOT NULL
    AND order_purchase_timestamp IS NOT NULL
GROUP BY YEAR(order_purchase_timestamp), MONTH(order_purchase_timestamp)
ORDER BY ano, mes;
```


> Nova hipotese: quais pedidos não foram entregues e são pra quais cidade/estado?

```sql
SELECT c.customer_city AS Cidade, c.customer_state AS Estado, p.order_status AS Status_do_pedido
FROM olist_dataset.default.clientes_olist C 
INNER JOIN olist_dataset.default.pedidos_olist P ON C.customer_id = P.customer_id
WHERE p.order_status != 'delivered'
ORDER BY c.customer_state, c.customer_city, p.order_status
```
